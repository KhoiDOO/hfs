```python
import PIL
import requests
import torch
import numpy as np
import matplotlib.pyplot as plt
from io import BytesIO

from diffusers import StableDiffusionInstructPix2PixPipeline


def download_image(url):
    response = requests.get(url)
    return PIL.Image.open(BytesIO(response.content)).convert("RGB")
```

```python
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    "timbrooks/instruct-pix2pix", torch_dtype=torch.float16
)
device = torch.device('cuda', index=1)
# pipe = pipe.to("cuda")
pipe = pipe.to(device)
```

```python
sample_path = '/media/mountHDD2/git/hfs/sample.jpg'
sample_img = PIL.Image.open(sample_path).convert("RGB").resize((512, 512))

plt.imshow(np.array(sample_img))
```

```python
prompt = "make the image look like it is captured from a phone image"
image = pipe(prompt=prompt, image=sample_img).images[0]

plt.imshow(np.array(image))
```
